```python
import requests
```

```python
f = open("links.txt")
data = f.read()
f.close()
pages = data.split("\n")
pages
```

```python
base = 'https://simple.wikipedia.org'
for page in pages:
    url = base + page
    print(url)
    resp = requests.get(url)
    f = open(url.split("/")[-1] + ".html", "w")
    f.write(resp.text)
    f.close()
    print(resp)
```
