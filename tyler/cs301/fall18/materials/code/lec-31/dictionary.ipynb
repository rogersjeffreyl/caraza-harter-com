```python
import pandas as pd
import sqlite3
import json
```

```python
with open("dictionary.json") as f:
    d = json.load(f)
```

```python
words = []
defs = []
for k,v in d.items():
    words.append(k)
    defs.append(v)
table = pd.DataFrame({"word":words, "definition":defs})
table
```

```python
conn = sqlite3.connect("vocab.db")
table.to_sql(con=conn, name="definitions")
conn.close()
```
