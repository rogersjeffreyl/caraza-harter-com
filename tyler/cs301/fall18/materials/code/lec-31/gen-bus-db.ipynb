```python
import pandas as pd
```

```python
df = pd.read_csv("Metro_Transit_Ridership_by_Route_Weekday.csv")
df.head()
```

```python
table = df[["StopID", "Route", "Lat", "Lon", "DailyBoardings"]]
table
```

```python
import sqlite3
conn = sqlite3.connect('bus.db')
```

```python
table.to_sql("boarding", conn, if_exists="replace")
```

```python
pd.read_csv("Metro_Transit_Bus_Routes.csv").to_sql("routes", conn, if_exists="replace")
```

```python
cursor = conn.execute("select * from boarding")
for row in cursor:
    print(row)
```
