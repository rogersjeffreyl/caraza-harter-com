```python
from pandas import DataFrame, Series
import pandas as pd
```

```python
df = pd.read_csv('IMDB-Movie-Data.csv')
df.head()
```

```python
df['Revenue'] = df.Revenue.str.replace("M", "").astype(float)
```

```python
df.head()
```

```python
import sqlite3

c = sqlite3.connect('movies.db')
df[['Title', 'Director', 'Year', 'Runtime', 'Rating', 'Revenue']].to_sql("movies", c, if_exists='replace', index=False)
c.close()
```

```python
c = sqlite3.connect('movies.db')
df = pd.read_sql("select * from movies", c)
c.close()
df
```
