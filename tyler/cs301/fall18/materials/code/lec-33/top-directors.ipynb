```python
import sqlite3
from pandas import DataFrame, Series
import pandas as pd
```

```python
c = sqlite3.connect("movies.db")
```

```python
df = pd.read_sql("SELECT * FROM movies", c)
df.head()
```

```python
groups = {}
for t in df.iterrows():
    row = t[1]
    director = row['Director']
    if not director in groups:
        groups[director] = []
    groups[director].append(row)
```

```python
for director in groups:
    rows = groups[director]
    ratings_sum = 0
    for row in rows:
        ratings_sum += row['Rating']
    avg = ratings_sum / len(rows)
    print(director, avg)
```
