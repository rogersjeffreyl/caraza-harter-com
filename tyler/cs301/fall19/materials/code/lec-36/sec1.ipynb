```python
import matplotlib
from matplotlib import pyplot as plt
```

```python
matplotlib.rcParams["font.size"] = 18
```

```python
ax = (s/1e6).plot.barh(figsize=(5, 1.5), color="0.5")
ax.set_xlabel("Dollars (Millions)")
ax.set_title("Annual Spending")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
```

```python
def get_ax(figsize=(5, 1.5)):
    _, ax = plt.subplots(figsize=figsize)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    return ax

ax = get_ax()
(s/1e6).plot.barh(color="0.5", ax=ax)
ax.set_xlabel("Dollars (Millions)")
ax.set_title("Annual Spending")
```

```python
# guesses: 80, 6, 3, 2, 1
df = pd.read_sql("""
SELECT Route, SUM(DailyBoardings) AS daily
FROM boarding
GROUP BY Route
ORDER BY daily DESC
""", conn)
df = df.set_index("Route")
df.head(8)
```

```python
s = df["daily"].iloc[:5]
s["other"] = df["daily"].iloc[5:].sum()
s
```
