```python
import requests
```

```python
baseurl = "https://tyler.caraza-harter.com/cs301/spring19/materials/code/lec-30/data/"
url = baseurl + "state_files.txt"
```

```python
r = requests.get(url)
r
```

```python
type(r)
```

```python
# assert(r.status_code == 200)
r.raise_for_status()
```

```python
states = r.text.split("\n")
states
```

```python
# key:col name, value: list of values in that column
table = {"state":[], "2000":[], "2010":[], "2015":[]} 

for state in states:
    url = baseurl + state
    print("DOWNLOAD", baseurl + state)
    r = requests.get(url)
    r.raise_for_status()
    row = r.json()
    for key in row:
        column = table[key] # column list
        column.append(row[key])
    table["state"].append(state.replace(".json", ""))
```
