```python
import requests
from bs4 import BeautifulSoup
```

```python
# STAGE 1: grab the URLs
```

```python
url = "https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States"
resp = requests.get(url)
resp
```

```python
resp.raise_for_status()
```

```python
html = resp.text
doc = BeautifulSoup(html, "html.parser")
```

```python
print(doc.prettify())
```

```python
tables = doc.find_all("table")
```

```python
len(tables)
```

```python
table = None

for t in tables:
    rows = t.find_all("tr")
    if len(rows) >= 50:
        table = t
        break
        
assert(table != None)
```

```python
trs = table.find_all("tr")
len(trs)
```

```python
urls = []
for tr in trs[2:]:
    th = tr.find("th")
    a = th.find("a")
    urls.append("https://en.wikipedia.org" + a.attrs["href"])
urls
```

```python
# STAGE 2: download all the pages
```

```python
for url in urls:
    name = url.split("/")[-1]
    name = name.split("_(")[0]
    print(name)
    
    resp = requests.get(url)
    resp.raise_for_status()
    
    f = open(name+".html", "w")
    f.write(resp.text)
    f.close()
```

```python
# STAGE 3: get stats from the HTML files
```

```python
def extract_stats(path):
    f = open(path)
    html = f.read()
    f.close()
    doc = BeautifulSoup(html, "html.parser")
    stats = {}
    for tr in doc.find_all("tr"):
        cells = tr.find_all(["th", "td"])
        if len(cells) == 2:
            key = cells[0].get_text()
            val = cells[1].get_text()
            stats[key] = val
    return stats
```

```python
wi = extract_stats("Wisconsin.html")
wi["Capital"]
```
