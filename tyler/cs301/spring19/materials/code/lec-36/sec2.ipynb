```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
cols = ["sep-len", "sep-wid", "pet-len", "pet-wid", "class"]
df = pd.read_csv(url, names=cols)
df.head()
```

```python
df["class"].value_counts() # GROUP BY, COUNT(*)
```

```python
iris = df[df["class"] == "Iris-versicolor"]
iris.plot.scatter(x="pet-wid", y="pet-len", title="versicolor")
iris = df[df["class"] == "Iris-virginica"]
iris.plot.scatter(x="pet-wid", y="pet-len", title="virginica")
iris = df[df["class"] == "Iris-setosa"]
ax = iris.plot.scatter(x="pet-wid", y="pet-len", title="setosa")
type(ax)
```

```python
iris = df[df["class"] == "Iris-versicolor"]
ax = iris.plot.scatter(x="pet-wid", y="pet-len", label="versicolor",
                       title="Flowers", color="red", marker="o")

iris = df[df["class"] == "Iris-virginica"]
iris.plot.scatter(x="pet-wid", y="pet-len", label="virginica",
                  ax=ax, color="blue", marker="v")

iris = df[df["class"] == "Iris-setosa"]
iris.plot.scatter(x="pet-wid", y="pet-len", label="setosa",
                  ax=ax, color="green", marker="^")
```

```python
varieties = list(df["class"].value_counts().index)
varieties
```

```python
ax = None

colors = ["red", "blue", "green"]
markers = ["+", "*", "1"]

for variety in varieties:
    iris = df[df["class"] == variety]
    ax = iris.plot.scatter(x="pet-wid", y="pet-len",
                           label=variety,
                           title="Flowers", color=colors.pop(0),
                           marker=markers.pop(0), ax=ax)
```
