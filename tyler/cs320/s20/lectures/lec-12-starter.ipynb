```python
from graphviz import Graph, Digraph
from IPython.core.display import display
```

```python
# Binary Tree
class Node:
    def __init__(self, val):
        self.val = val
        # binary tree
        self.left = None
        self.right = None
        
    def dump(self):
        print(self.val)
        if self.left != None:
            self.left.dump()
        if self.right != None:
            self.right.dump()
            
    def _repr_svg_(self, g=None):
        if g == None:
            gv = Digraph()
        else:
            gv = g
        gv.node(self.val)
        
        if self.left != None:
            self.left._repr_svg_(gv)
            gv.edge(self.val, self.left.val, label="L")
        if self.right != None:
            self.right._repr_svg_(gv)
            gv.edge(self.val, self.right.val, label="R")
            
        return gv._repr_svg_()
    
    def contains(self, val):
        display("does the tree contain "+val+"?")
        display(self)
        
        if self.val == val:
            return True
        
        if self.left != None and self.left.contains(val):
            return True
        
        if self.right != None and self.right.contains(val):
            return True
        
        return False
    
        
root = Node("A")
root.left = Node("B")
root.right = Node("C")
root.left.right = Node("D")
root.left.left = Node("E")
root.contains("Z")
```
